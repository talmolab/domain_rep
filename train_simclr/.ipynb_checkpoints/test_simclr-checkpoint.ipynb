### Create dataloader pipeline

```python
class ImageNet64(torch.utils.data.Dataset):
    def __init__(self,xs,ys,mode):
        self.mode = mode
        self.dataset = torch.from_numpy(xs)
        self.size = len(xs)
        self.labels = torch.tensor(ys)
        if self.mode == 'train':
            self.transform = torchvision.transforms.Compose([torchvision.transforms.RandomResizedCrop(64), torchvision.transforms.RandomHorizontalFlip(0.5)])
        elif self.mode == 'val': 
            self.transform = None
        else:
            print('Mode must be either `train` or `val`')
                                                              
    def __len__(self):
        return self.size
    def  __getitem__(self, index):
        if self.mode == 'train':
            return self.transform(self.dataset[index]), self.labels[index]
        else:
            return self.dataset[index], self.labels[index]
```

## reload trained model

```python
class SimCLRModel(pl.LightningModule):
    def __init__(self):
        super().__init__()

        AlexNet = alexnet()
        self.backbone = torch.nn.Sequential(*list(AlexNet.children())[:-1])
        # print(self.backbone)
        hidden_dim = 9216
        self.projection_head = SimCLRProjectionHead(hidden_dim, hidden_dim, 128)

        self.criterion = NTXentLoss()

    def forward(self, x):
        h=self.backbone(x).flatten(start_dim=1)
        z = self.projection_head(h)
        return z

    def training_step(self, batch, batch_idx):
        (x0, x1), _, _ = batch
        z0 = self.forward(x0)
        z1 = self.forward(x1)
        loss = self.criterion(z0, z1)
        self.log("train_loss_ssl", loss)
        return loss

    def configure_optimizers(self):
        optim = torch.optim.Adam(params=self.parameters())
        scheduler = torch.optim.lr_scheduler.CosineAnnealingLR(
            optim, 200, eta_min = 0
        )
        return [optim], [scheduler]
```

## Train Linear Probe

```python
import torchmetrics
class SimCLRTester(pl.LightningModule):
    def __init__ (self, input_shape, SimCLR_model, num_classes, learning_rate=0.3):
        super().__init__()
        self.save_hyperparameters()
        self.learning_rate = learning_rate
        self.num_classes=num_classes
        
        self.encoder = SimCLR_model.backbone #alexnet backbone
        # freeze params
        self.encoder.eval()
        for param in self.encoder.parameters():
            param.requires_grad = False
        n_sizes = self._get_conv_output(input_shape)
        self.classifier = torch.nn.Linear(9216, self.num_classes)

    # returns the size of the output tensor going into Linear layer from the conv block.
        
    def _get_conv_output(self, shape):
        batch_size = 1
        tmp_input = torch.autograd.Variable(torch.rand(batch_size, *shape))
        output_feat = self._forward_features(tmp_input) 
        n_size = output_feat.data.view(batch_size, -1).size(1)
        return n_size

    # returns the feature tensor from the conv block
    def _forward_features(self, x):
        x = self.encoder(x)
        return x

    # will be used during inference
    def forward(self, x):
        x = self._forward_features(x).flatten(1)
        pred = self.classifier(x)
        return pred
        

    def training_step(self, batch, batch_idx):
        x, y = batch
        logits = self(x)
        #y = torch.squeeze(y,1)
        # y = y.unsqueeze(1)
        # print(x.shape)
        # print(logits.shape)
        # print(y.shape)
        loss = torch.nn.functional.cross_entropy(logits, y)
        # training metrics
        # preds = torch.argmax(logits, dim=1)
        acc = torchmetrics.functional.accuracy(logits, y)
        self.log('train_loss', loss, on_step=True, on_epoch=True, logger=True)
        self.log('train_acc', acc, on_step=True, on_epoch=True, logger=True)

        return loss

    def validation_step(self, batch, batch_idx):
        x, y = batch
        logits = self(x)
        # y = torch.squeeze(y,1)
        # y = y.unsqueeze(1)
        # print(x.shape)
        # print(logits.shape)
        # print(y.shape)
        loss = torch.nn.functional.cross_entropy(logits, y)

        # validation metrics
        # preds = torch.argmax(logits, dim=1)
        acc = torchmetrics.functional.accuracy(logits, y)
        self.log('val_loss', loss, prog_bar=True)
        self.log('val_acc', acc, prog_bar=True)
        return loss

    def test_step(self, batch, batch_idx):
        x, y = batch
        logits = self(x)
        # y = torch.squeeze(y,1)
        # y = y.unsqueeze(1)
        # print(x.shape)
        # print(logits.shape)
        # print(y.shape)
        loss = torch.nn.functional.cross_entropy(logits, y)
        # validation metrics
        # preds = torch.argmax(logits, dim=1)
        acc = torchmetrics.functional.accuracy(logits, y)
        self.log('test_loss', loss, prog_bar=True)
        self.log('test_acc', acc, prog_bar=True)
        return loss

    def configure_optimizers(self):
        optimizer = torch.optim.SGD(self.parameters(),self.learning_rate,0.9,1e-9)
        scheduler = torch.optim.lr_scheduler.StepLR(optimizer,step_size=30)
        #LARS(params=self.parameters(),lr=self.learning_rate,momentum=0.9,weight_decay=1e-6)
        return [optimizer]#,[scheduler]
```
